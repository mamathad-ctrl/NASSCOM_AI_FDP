```python
##EXERCISE 1 ON vectors
import numpy as np
```

```python
v1=np.array([78,89,90])
v2=np.array([78,79,90])
v3=np.array([98,89,40])
AB = np.linalg.norm(v1 - v2) #L2 norms
AC = np.linalg.norm(v1 - v3)
BC = np.linalg.norm(v2 - v3)
```

```python
print(AB)
print(BC)
```

```python
print(AC)
```

```python
print(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))
```

```python
print(np.dot(v2, v3) / (np.linalg.norm(v2) * np.linalg.norm(v3)))
```

```python
print(np.dot(v3, v1) / (np.linalg.norm(v3) * np.linalg.norm(v1)))
```

```python
import numpy as np

A = np.array([
    [2,1,3],
    [1,3,2],
    [3,2,1]
])

b = np.array([29,34,31])

x = np.linalg.solve(A,b)
print(x)
```

```python
print("A @ x =", A @ x)
print("b     =", b)
#print("Verified:", np.allclose(A @ x, b))
```
